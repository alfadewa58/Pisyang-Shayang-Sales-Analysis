## **Business Understanding**

**Context**

Pisyang Shayang adalah perusahaan di bidang food and beverages (F&B) yang fokus menjual **pisang goreng** dengan berbagai pilihan topping kreatif. Dengan target pasar anak muda dan keluarga, perusahaan ini mengandalkan penjualan melalui platform online seperti Gojek, Shopee, dan Grab, serta penjualan offline.

**Problem Statement**

1. Bagaimana pola penjualan Pisyang Shayang sepanjang tahun?
   - Kapan periode dengan penjualan tertinggi dan terendah?
   - Apakah ada bulan dengan potensi tinggi yang belum dioptimalkan?
   - Bagaimana pertumbuhan penjualan dari waktu ke waktu?
2. Apakah kenaikan harga yang terjadi pada bulan Juli/Agustus 2024 berdampak signifikan pada revenue?
   - Bagaimana perbandingan revenue sebelum dan setelah kenaikan harga?
   - Apakah jumlah unit yang terjual berkurang secara signifikan setelah kenaikan harga?
3. Bagaimana diversifikasi saluran penjualan Pisyang Shayang?
   - Saluran mana yang memberikan kontribusi penjualan tertinggi?
   - Bagaimana proporsi penjualan berdasarkan ukuran (small vs large)?
   - Apakah proporsi penjualan ini konsisten setiap bulan atau ada variasi?

**Goals**

1. Meningkatkan Efisiensi dan Strategi Penjualan
   - Menentukan bulan dengan potensi penjualan tertinggi untuk promosi besar-besaran.
   - Mengurangi low season sales dengan strategi pemasaran yang tepat.
2. Menganalisis Dampak Kenaikan Harga
   - Memastikan kenaikan harga tidak menurunkan revenue secara signifikan.
3. Mengoptimalkan Diversifikasi Sales Channels
   - Mengidentifikasi saluran penjualan yang perlu difokuskan atau dikembangkan.
   - Memahami preferensi pelanggan berdasarkan ukuran produk untuk merancang stok dan promosi yang tepat.

**Analysis Components**
1. Sales (pcs)
   - Sales Tertinggi dan Terendah:
     - Identifikasi bulan dengan penjualan terbanyak dan tersedikit.
   - High Potential Month & Low Potential Month:
     - Analisis potensi bulan untuk menentukan strategi promosi.
   - Sales Growth:
     - Hitung pertumbuhan penjualan dari bulan ke bulan atau tahun ke tahun.
2. Revenue Analysis
Premis: Pada bulan Juli/Agustus 2024, terjadi kenaikan harga produk. Analisis ini bertujuan untuk memahami dampaknya:
     - Bandingkan average/total sales di bulan sebelum dan setelah kenaikan harga untuk melihat apakah revenue meningkat meskipun unit sales berkurang.
     - Hitung jumlah unit sales yang dibutuhkan untuk mencapai revenue tertentu sebelum dan setelah kenaikan harga. Apakah kenaikan harga membuat perusahaan lebih efisien dalam menghasilkan revenue?
3. Diversifikasi Sales
   - Persentase Penjualan Berdasarkan Channels:
     - Identifikasi kontribusi penjualan dari Gojek, Shopee, Grab, QRIS, dan offline setiap bulan, kemudian hitung rata-ratanya.
   - Persentase Penjualan Berdasarkan Ukuran:
     - Analisis penjualan berdasarkan ukuran produk (small dan large). Apakah ada pergeseran tren ukuran favorit pelanggan?

```python
def analyze_features(df):
    # Create empty lists to store results
    feature_names = []
    missing_values = []
    missing_percentages = []
    unique_samples = []
    total_unique = []
    
    # Calculate metrics for each column
    for col in df.columns:
        # Get feature name
        feature_names.append(col)
        
        # Get missing value count
        missing = df[col].isnull().sum()
        missing_values.append(missing)
        
        # Calculate missing percentage
        missing_pct = (missing / len(df)) * 100
        missing_percentages.append(f"{missing_pct:.2f}%")
        
        # Get total unique count
        total_unique.append(len(df[col].unique()))
        
        # Get 5 unique samples
        unique_vals = df[col].unique()
        samples = unique_vals[:5] if len(unique_vals) >= 5 else unique_vals
        unique_samples.append(samples)
    
    # Create results dataframe
    results = pd.DataFrame({
        'Feature': feature_names,
        'Missing Values': missing_values,
        'Missing Percentage': missing_percentages,
        'Total Unique Values': total_unique,
        'Sample Unique Values': unique_samples
    })
    
    return results

# Display the analysis results
feature_analysis = analyze_features(transactions)
print("\nFeature Analysis:")
feature_analysis
```

### Outliers Handling

```python
# Function to detect outliers using IQR method
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound
```

```python
# Detect outliers for numerical columns
numerical_columns = ['Jumlah Produk Terjual (Unit)', 'Penjualan Kotor', 'Penjualan']

for column in numerical_columns:
    outliers, lower_bound, upper_bound = detect_outliers(transactions, column)
    print(f"\nOutliers in {column}:")
    print(f"Number of outliers: {len(outliers)}")
    print(f"Lower bound: {lower_bound}")
    print(f"Upper bound: {upper_bound}")
```

```python
# Perform statistical test to compare August 2023 vs 2024 sales
from scipy import stats
import numpy as np

# Get daily sales data for August 2023 and 2024
aug_2023_daily = transactions[
    (transactions['Year'] == 2023) & 
    (transactions['Month'] == 'Agustus')
]['Jumlah Produk Terjual (Unit)'].values

aug_2024_daily = transactions[
    (transactions['Year'] == 2024) & 
    (transactions['Month'] == 'Agustus')
]['Jumlah Produk Terjual (Unit)'].values

# First check normality using Shapiro-Wilk test
_, p_val_2023 = stats.shapiro(aug_2023_daily)
_, p_val_2024 = stats.shapiro(aug_2024_daily)

print("\nNormality Test Results:")
print("-" * 40)
print(f"August 2023 Shapiro-Wilk p-value: {p_val_2023:.4f}")
print(f"August 2024 Shapiro-Wilk p-value: {p_val_2024:.4f}")

# If p-value < 0.05, data is not normally distributed
if p_val_2023 < 0.05 or p_val_2024 < 0.05:
    print("\nData is not normally distributed, using Mann-Whitney U test instead of t-test")
    # Perform Mann-Whitney U test for non-normal distributions
    statistic, p_value = stats.mannwhitneyu(
        aug_2023_daily, aug_2024_daily,
        alternative='two-sided'
    )
    test_name = "Mann-Whitney U Test"
else:
    print("\nData is normally distributed, can use independent t-test")
    # Perform independent t-test for normal distributions
    statistic, p_value = stats.ttest_ind(
        aug_2023_daily, aug_2024_daily
    )
    test_name = "Independent t-test"

print(f"\n{test_name} Results:")
print("-" * 40)
print(f"Test statistic: {statistic:.4f}")
print(f"p-value: {p_value:.4f}")

# Interpret results
alpha = 0.05
print("\nInterpretation:")
if p_value < alpha:
    print(f"Since p-value ({p_value:.4f}) < {alpha}, there is a statistically significant")
    print("difference between August 2023 and August 2024 sales.")
else:
    print(f"Since p-value ({p_value:.4f}) >= {alpha}, there is NOT enough evidence")
    print("to conclude a significant difference between August 2023 and August 2024 sales.")

# Calculate effect size
if test_name == "Mann-Whitney U Test":
    n1 = len(aug_2023_daily)
    n2 = len(aug_2024_daily)
    z_score = stats.norm.ppf(p_value/2)
    effect_size = abs(z_score) / np.sqrt(n1 + n2)
else:
    # Cohen's d for t-test
    pooled_std = np.sqrt(((len(aug_2023_daily) - 1) * np.var(aug_2023_daily) + 
                         (len(aug_2024_daily) - 1) * np.var(aug_2024_daily)) / 
                        (len(aug_2023_daily) + len(aug_2024_daily) - 2))
    effect_size = abs(np.mean(aug_2023_daily) - np.mean(aug_2024_daily)) / pooled_std

print(f"\nEffect size: {effect_size:.4f}")
print("Effect size interpretation:")
if test_name == "Mann-Whitney U Test":
    print("- 0.1: Small effect")
    print("- 0.3: Medium effect") 
    print("- 0.5: Large effect")
else:
    print("- 0.2: Small effect")
    print("- 0.5: Medium effect")
    print("- 0.8: Large effect")
```

Setelah Agustus 2024 (bulan di mana harga naik) ternyata sales dan jumlah barang terjualnya semaik naik juga. Ini bisa terjadi karena:
1. Peningkatan Promosi atau Word-of-Mouth
- Kenaikan harga sering kali disertai dengan strategi promosi baru, seperti iklan, kolaborasi, atau diskon bundling.
- Selain itu, konsumen mungkin penasaran dengan "apa yang membuat produk ini naik harga" dan mencoba membelinya, yang dapat meningkatkan penjualan. Word-of-mouth juga bisa meningkat karena konsumen membicarakan produk yang mereka anggap naik kelas.

```python
# Calculate monthly units sold for each size category
monthly_size_units = transactions_standardized.groupby(['Year', 'Month', 'Kategori'])['Jumlah Produk Terjual (Unit)'].sum().reset_index()

# Create a function to categorize products
def categorize_size(category):
    if category == 'Small':
        return 'Small'
    elif 'Large' in category:  # Changed to catch any category containing 'Large'
        return 'Large'
    else:
        return 'Other'

# Add size category column
monthly_size_units['Size_Category'] = monthly_size_units['Kategori'].apply(categorize_size)

# Print unique categories and their mappings to debug
print("\nCategory Mappings:")
for cat in monthly_size_units['Kategori'].unique():
    print(f"{cat} -> {categorize_size(cat)}")

# Create pivot table with explicit categories
size_summary = monthly_size_units.groupby(['Year', 'Month', 'Size_Category'])['Jumlah Produk Terjual (Unit)'].sum().reset_index()

# Create pivot table without categorical type
size_pivot = pd.pivot_table(size_summary,
                          values='Jumlah Produk Terjual (Unit)', 
                          index=['Year', 'Month'],
                          columns='Size_Category',
                          fill_value=0).reset_index()

# Ensure all size categories exist
for category in ['Small', 'Large', 'Other']:
    if category not in size_pivot.columns:
        size_pivot[category] = 0

# Calculate total quantity per month
size_pivot['Total_Quantity'] = size_pivot['Small'] + size_pivot['Large'] + size_pivot['Other'] 

# Calculate percentages
size_pivot['Small_Percentage'] = (size_pivot['Small'] / size_pivot['Total_Quantity'] * 100).round(2)
size_pivot['Large_Percentage'] = (size_pivot['Large'] / size_pivot['Total_Quantity'] * 100).round(2)
size_pivot['Other_Percentage'] = (size_pivot['Other'] / size_pivot['Total_Quantity'] * 100).round(2)

# Display the first few rows
print("\nMonthly Size Distribution:")
print(size_pivot)
```
